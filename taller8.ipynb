```python
nltk.download('vader_lexicon')

sa = SentimentIntensityAnalyzer()
```

```python
frase = "¡Hoy es un dia maravilloso! Estoy muy feliz y emocionado por todo lo que me depara el dia hoy"
sentimiento = sa.polarity_scores(frase)
```

```python
if sentimiento['compound'] > 0:
  print('La frase es positiva')
elif sentimiento['compound'] < 0:
  print('La frase es negativa')
else:
  print('La frase es neutral')

print('Puntajes de sentimientos: ', sentimiento)
```

```python
import re
from nltk.corpus import stopwords

def clean_text(text):
    text = text.lower()

    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)

    tokens = nltk.word_tokenize(text)

    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words and len(token) > 4]

    cleaned_text = " ".join(tokens)
    return cleaned_text

cleaned_texto = clean_text(book)
cleaned_texto
```

```python
stop_words = sorted(nltk.corpus.stopwords.words('english'))

tokens = nltk.word_tokenize(cleaned_texto)

relevant = [token for token in tokens if token not in stop_words]
```

```python
relevant
```

```python
freq = nltk.FreqDist(relevant)

top_words = freq.most_common(30)

top_words
```
