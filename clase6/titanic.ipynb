```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#cargamos los datos desde una url
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
```

```python
#usando pandas lo ponemos en un dataframe
titanic = pd.read_csv(url)

#mostramos los primeros elementos
titanic.head()
```

```python
#Configuramos el PassengerId como índice
titanic.set_index('PassengerId').head()
```

```python
# sumamos las veces que se repiten hombres y mujeres
#lo guardamos en una variable cada uno

males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

# ponemos hombres y mujeres en una lista
proportions = [males, females]

# Creamos el pie chart
plt.pie(
    # usamos la lista creada
    proportions,
    
    # ponemos etiquetas
    labels = ['Males', 'Females'],
    
    # quitamos las sombras
    shadow = False,
    
    # asignamos colores
    colors = ['blue','red'],
    
    # sacamos el un slide
    explode = (0.15 , 0),
    
    # asignamos un águlo de 90
    startangle = 90,
    
    # mostramo el porcentaje con dos decimales
    autopct = '%1.2f%%'
    )

# seteamos los ejes como iguales
plt.axis('equal')

# Ponemos las etiquetas
plt.title("Sex Proportion")

# Mostramos el pu
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que ya tienes el DataFrame titanic cargado
# Crear el gráfico de dispersión
scatter_plot = sns.scatterplot(x='Age', y='Fare', data=titanic, hue='Sex')

# Añadir el título
scatter_plot.set_title('Fare x Age')

# Ajustar los ejes
scatter_plot.set_ylim(-5, None)
scatter_plot.set_xlim(-5, 85)

# Mostrar el gráfico
plt.show()
```

```python
# ordenamos los valoresdeel mayor al menor
df = titanic.Fare.sort_values(ascending = False)
df

# creamos los bins con numpy
binsVal = np.arange(0,600,10)
binsVal

# creamos la gráfica
plt.hist(df, bins = binsVal)

# Ponemos títuloos y eetiquetas
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed Histrogram')

# mostramos la gráfica
plt.show()
```

```python
live = (titanic['Survived'] == 1).sum()
dead = (titanic['Survived'] == 0).sum()

# ponemos hombres y mujeres en una lista
proportions = [live, dead]

# Creamos el pie chart
plt.pie(
    # usamos la lista creada
    proportions,
    
    # ponemos etiquetas
    labels = ['Vivo', 'Muerto'],
    
    # quitamos las sombras
    shadow = False,
    
    # asignamos colores
    colors = ['green','gray'],
    
    # sacamos el un slide
    explode = (0.15 , 0),
    
    # asignamos un águlo de 90
    startangle = 90,
    
    # mostramo el porcentaje con dos decimales
    autopct = '%1.2f%%'
    )

# seteamos los ejes como iguales
plt.axis('equal')

# Ponemos las etiquetas
plt.title("Lived Propagation")

# Mostramos el pu
plt.tight_layout()
plt.show()
```
