```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

```python
CS_Players = pd.read_csv('csgo_player_stats.csv')
CS_Players.head()
```

```python
CS_Players.tail()
```

```python
CS_Players.info()
```

```python
CS_Players.isnull().sum()
```

```python
top10_players = CS_Players.nlargest(10, ' Total Kills')

total_kills = top10_players[' Total Kills']
total_deaths = top10_players[' Total Deaths']

n = len(total_kills)

x = np.arange(n)
pt.bar(x, total_kills, facecolor='#9999ff', edgecolor='white')
pt.bar(x, -total_deaths, facecolor='#ff9999', edgecolor='white')

pt.xlabel('Jugadores')
pt.ylabel('Total Kills y Total Deaths')
pt.title('Top 10 Jugadores con Ms Kills y sus Deaths')

pt.xticks(x, top10_players['Name'], rotation=45, ha='right')

pt.show()

print('Este grafico demuestra la relacion entre la cantidad de killes realizadas contra la cantidad de muertes recibidas\n')
print('Se puede observar que los que mas se arriesgan a conseguir una kill, mueren mas seguido')
```

```python
top10_players = CS_Players.nsmallest(10, ' Total Kills')

total_kills = top10_players[' Total Kills']
total_deaths = top10_players[' Total Deaths']

n = len(total_kills)

x = np.arange(n)
pt.bar(x, total_kills, facecolor='#9999ff', edgecolor='white')
pt.bar(x, -total_deaths, facecolor='#ff9999', edgecolor='white')

pt.xlabel('Jugadores')
pt.ylabel('Total Kills y Total Deaths')
pt.title('Top 10 Jugadores con menos Kills y sus Deaths')

pt.xticks(x, top10_players['Name'], rotation=45, ha='right')

pt.show()

print('Este grafico muestra a los jugadores con menos kills y sus muertes')
print('Demostrando que a a pesar de no conseguir muchas kills, la cantidad de sus muertes es mayor')
```

```python
top5_KD = CS_Players.nlargest(10, ' Kill/Death Ratio')

labels = top5_KD['Name']
sizes = top5_KD[' Kill/Death Ratio']

fig, ax = pt.subplots()
ax.bar(labels, sizes)

ax.set_xlabel('Jugadores')
ax.set_ylabel('Kill/Death Ratio')
ax.set_title('Top 5 Juegadores con mejor K/D')

pt.xticks(rotation=45, ha='right')

pt.show()

print('En base a esto se puede medir el K/D, con esto se puede definir cuales son los mejores jugadores')
print('Ya que se muestra a los jugadores con mas kills segun sus muertes')
```

```python
top5_KD = CS_Players.nsmallest(10, ' Kill/Death Ratio')

labels = top5_KD['Name']
sizes = top5_KD[' Kill/Death Ratio']

fig, ax = pt.subplots()
ax.bar(labels, sizes)

ax.set_xlabel('Jugadores')
ax.set_ylabel('Kill/Death Ratio')
ax.set_title('Top 5 Juegadores con peor K/D')

pt.xticks(rotation=45, ha='right')

pt.show()

print('Este grafico es la contraparte del anterior, mostrardo que jugadores tiene peor K/D')
print('Estos serian los jugadores que menos kills han hecho segun las muertes que tuvieron')
```

```python
top10_players = CS_Players.nsmallest(10, 'Damage Per Round')

total_ADR = top10_players['Damage Per Round']
total_KD = top10_players[' Kill/Death Ratio']

total_ADR = total_ADR/100

n = len(total_ADR)

x = np.arange(n)
pt.bar(x, total_ADR, facecolor='#9999ff', edgecolor='white')
pt.bar(x, -total_KD, facecolor='#ff9999', edgecolor='white')

pt.xlabel('Jugadores')
pt.ylabel('Total Kills y Total Deaths')
pt.title('Top 10 Jugadores con menor ADR y sus Kills')

pt.xticks(x, top10_players['Name'], rotation=45, ha='right')

pt.show()

print('Este grafico muestra la relacion entre el K/D de los jugadores con respecto al dao realizado por ronda')
print('El dao por ronda se mide en numeros del 1 al 100 pero para que se pueda ver mejor el grafico lo vamos a medir del 0 al 1')
```
