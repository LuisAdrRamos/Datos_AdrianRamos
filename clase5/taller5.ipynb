```python
def rating_funcion(x):
    if x >= 8.0:
        return "MOTY (Movie of the year)"
    elif  x >= 7.0:
        return "Good"
    elif x >= 6.0:
        return "Normal"
    else:
        return "Bad"
```

```python
movies_df["rating_category"] = movies_df["Rating"].apply(rating_funcion)
```

```python
movies_df.head(10)
```

```python
def recaudacion(x):
    if x >= 20:
        return "Exito Total"
    elif x >= 10:
        return "Exito Parcial"
    else:
        return "Fracaso Total"
```

```python
movies_df["revenue"] = movies_df["Revenue (Millions)"].apply(recaudacion)
```
